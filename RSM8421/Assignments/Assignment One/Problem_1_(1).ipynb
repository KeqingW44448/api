```python
from tensorflow import keras
import numpy as np
#Every one finish this already
from keras.models import Sequential
from keras.layers import Input, Dense, Activation
from keras.datasets import mnist
from keras.optimizers import Adam
from keras.utils import to_categorical # Helper Function,
from keras.callbacks import TensorBoard # Supervised Learning Curve, real time monitoring , showing the process of the learning of the neural network,
from keras.datasets import boston_housing
from sklearn.preprocessing import StandardScaler
```

## Exercise 1

```python
# Load the dataset
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
scalar = StandardScaler()
x_train2 = scalar.fit_transform(x_train) # Why we need standard scaler for the model fitting
x_test2 = scalar.transform(x_test)

print("Training Dataset Size: ", x_train2.shape)
print("Training Labels Size: ", y_train.shape)
print("Testing Dataset Size: ", x_test2.shape)
print("Testing Labels Size: ", y_test.shape)

model1 = Sequential()
model1.add(Input(shape=(13,)))
model1.add(Dense(15, activation="relu"))
model1.add(Dense(15, activation="relu"))
model1.add(Dense(15, activation="relu"))
model1.add(Dense(1, activation="relu"))

model1.summary()
```

```python
model1.compile(loss='mae', optimizer=Adam(learning_rate=0.01))
callbacks = [
    TensorBoard(log_dir='./exercise1', histogram_freq=1),
]
history = model1.fit(x_train2, y_train, batch_size=10, epochs=20, validation_split=0.15,
                     callbacks=callbacks)
```

```python
model1.evaluate(x_test2, y_test)
```
