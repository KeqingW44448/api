# Second Step: Regulization: L1 (|x|) and L2 (x^2)
##Add L1 & L2 Regularization Grid Search with alpha values: [0.0001, 0.001, 0.01, 0.1] to select the optimal mae_test

```python
from tensorflow import keras
import numpy as np
from keras.models import Sequential
from keras.layers import Input, Dense
from keras.optimizers import Adam
from keras.regularizers import l1, l2
from keras.datasets import boston_housing
from sklearn.preprocessing import StandardScaler

# -----------------------------
# Load and preprocess data
# -----------------------------
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()

np.random.seed(0)
index = np.arange(len(x_train))

# Take a biased training set - top 100 house prices
index = sorted(index, key=lambda x: -y_train[x])
x_train = x_train[index]
y_train = y_train[index]

sample = 100
x_train, y_train = x_train[:sample], y_train[:sample]

# Scale features
scaler = StandardScaler()
x_train2 = scaler.fit_transform(x_train)
x_test2 = scaler.transform(x_test)

print("Training Dataset Size:", x_train2.shape)
print("Testing Dataset Size:", x_test2.shape)

# -----------------------------
# Grid search for L1 and L2 alpha
# -----------------------------
alphas = [1e-4, 1e-3, 1e-2, 0.1]
results_l1 = {}
results_l2 = {}

for alpha in alphas:
    # L1 model
    model_l1 = Sequential([
        Input(shape=(13,)),
        Dense(100, activation='relu', kernel_regularizer=l1(alpha)),
        Dense(100, activation='relu', kernel_regularizer=l1(alpha)),
        Dense(100, activation='relu', kernel_regularizer=l1(alpha)),
        Dense(1)
    ])
    model_l1.compile(loss='mae', optimizer=Adam(learning_rate=0.01))
    model_l1.fit(x_train2, y_train, batch_size=sample, epochs=100, verbose=0)
    mae_l1 = model_l1.evaluate(x_test2, y_test, verbose=0)
    results_l1[alpha] = mae_l1
    print(f"L1 alpha={alpha}: MAE={mae_l1:.4f}")

    # L2 model
    model_l2 = Sequential([
        Input(shape=(13,)),
        Dense(100, activation='relu', kernel_regularizer=l2(alpha)),
        Dense(100, activation='relu', kernel_regularizer=l2(alpha)),
        Dense(100, activation='relu', kernel_regularizer=l2(alpha)),
        Dense(1)
    ])
    model_l2.compile(loss='mae', optimizer=Adam(learning_rate=0.01))
    model_l2.fit(x_train2, y_train, batch_size=sample, epochs=100, verbose=0)
    mae_l2 = model_l2.evaluate(x_test2, y_test, verbose=0)
    results_l2[alpha] = mae_l2
    print(f"L2 alpha={alpha}: MAE={mae_l2:.4f}")


# Find best alpha for L1 and L2

best_alpha_l1 = min(results_l1, key=results_l1.get)
best_alpha_l2 = min(results_l2, key=results_l2.get)

print(f"\nBest L1 alpha: {best_alpha_l1} with MAE={results_l1[best_alpha_l1]:.4f}")
print(f"Best L2 alpha: {best_alpha_l2} with MAE={results_l2[best_alpha_l2]:.4f}")
```

## Conclusion: The regulization for L2-parameter regulization with the weight alpha = 0.1 is the optimal hyperparameter structure with the minimum MAE = 10.0345.

 Explanation: Using the same one hundred epochs and the same training set, apply regularization (without altering the training set) to improve test set accuracy. The regularization methods I used, including the parameteres I tried are L1 - and L2 - Parameter Regularization with alphas = [1e-4, 1e-3, 1e-2, 0.1].

## Final Step: Dropout Regularization Grid Search with p = 0.1, 0.3, 0.5, 0.7

```python
from tensorflow import keras
import numpy as np
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
from keras.optimizers import Adam
from keras.datasets import boston_housing
from sklearn.preprocessing import StandardScaler


# Load and preprocess data
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()

np.random.seed(0)
index = np.arange(len(x_train))

# Take a biased training set - top 100 house prices
index = sorted(index, key=lambda x: -y_train[x])
x_train = x_train[index]
y_train = y_train[index]

sample = 100
x_train, y_train = x_train[:sample], y_train[:sample]

# Scale features
scaler = StandardScaler()
x_train2 = scaler.fit_transform(x_train)
x_test2 = scaler.transform(x_test)

print("Training Dataset Size:", x_train2.shape)
print("Testing Dataset Size:", x_test2.shape)

# Grid search for Dropout rate
dropout_rates = [0.1, 0.3, 0.5, 0.7]
results = {}

for rate in dropout_rates:
    model = Sequential([
        Input(shape=(13,)),
        Dense(100, activation='relu'),
        Dropout(rate),
        Dense(100, activation='relu'),
        Dropout(rate),
        Dense(100, activation='relu'),
        Dropout(rate),
        Dense(1)
    ])

    model.compile(loss='mae', optimizer=Adam(learning_rate=0.01))
    model.fit(x_train2, y_train, batch_size=sample, epochs=100, verbose=0)
    mae = model.evaluate(x_test2, y_test, verbose=0)
    results[rate] = mae
    print(f"Dropout rate={rate}: MAE={mae:.4f}")

# Find best dropout rate
best_rate = min(results, key=results.get)
print(f"\nBest Dropout rate: {best_rate} with MAE={results[best_rate]:.4f}")
```

## Conclusion: The regulization for dropout randomly ignore fraction p = 0.7 of neurons is the optimal hyperparameter structure with the minimum MAE = 9.9479

 Explanation: Using the same one hundred epochs and the same training set, apply regularization (without altering the training set) to improve test set accuracy. The regularization methods I used, including the parameteres I tried are the dropout with p = [0.1, 0.3, 0.5, 0.7].

Drop out works the best here, as the sample size is sufficicently large and the numbers of the important features and attributions are very large, therefore, the drop out will perform as the optimal result with test MSE = 9.9479.
